#Distribuição amostral e intervalo de confiança

```python
salarios = dados['SALARIO']
salarios
```

```python
media_amostral= np.mean(salarios)
media_amostral
```

```python
desvio_amostral = np.std(salarios)
desvio_amostral
```

```python
nível_confiança = 0.95
```

```python
tamanho_amostra = len(salarios)
tamanho_amostra
```

```python
from scipy import stats
```

```python
erro_padrão = stats.sem(salarios)
erro_padrão
```

```python
intervalo_confiança = stats.t.interval(nível_confiança, tamanho_amostra-1, loc=media_amostral, scale=erro_padrão)
```

```python
intervalo_confiança
```

#Featuring Engineering

```python
def preencher_nivel(gestor,nivel):
    if gestor==1:
       return "Pessoa Gestora"
    else:
       return nivel
```

```python
dados['NOVO_NIVEL']= dados.apply(lambda x: preencher_nivel(x['GESTOR?'], x['NIVEL']), axis=1)
```

#Gerando variáveis discretas

```python
def determinar_geracao(idade):
  if 39 < idade <=58:
    return "Geração x"
  elif 29 < idade <=39:
    return "Millenial"
  elif 13 < idade <=29:
    return "Geração z"
  else:
    return "Outra geração"
```

#Criando uma nova coluna

```python
dados['GERACAO'] = dados['IDADE'].apply(determinar_geracao)
```

```python
dados['GERACAO'].value_counts()
```
